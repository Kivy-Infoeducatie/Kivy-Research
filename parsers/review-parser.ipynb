```python
import pandas as pd
```

```python
df = pd.read_csv('../data/reviews.csv')

df
```

```python
df = df[['ReviewId', 'RecipeId', 'AuthorName', 'Rating', 'Review', 'DateSubmitted']]

df = df.rename(columns={
    'ReviewId': 'id',
    'RecipeId': 'recipe_id',
    'AuthorName': 'author_name',
    'Rating': 'rating',
    'Review': 'content',
    'DateSubmitted': 'created_at'
})

df.dropna(inplace=True)

df
```

```python
df.to_csv('output/parsed_reviews.csv', index=False)
```
